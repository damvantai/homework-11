```python
def compare_two_words(word1, word2):
    if (word1[0] != word2[0] and word1[1] == word2[1] and word1[2] == word2[2] and word1[3] == word2[3] and word1[4] == word2[4]):
           Dic_Edge[word1, word2] = 1
    if (word1[0] == word2[0] and word1[1] != word2[1] and word1[2] == word2[2] and word1[3] == word2[3] and word1[4] == word2[4]):
           Dic_Edge[word1, word2] = 1
    if (word1[0] == word2[0] and word1[1] == word2[1] and word1[2] != word2[2] and word1[3] == word2[3] and word1[4] == word2[4]):
           Dic_Edge[word1, word2] = 1
    if (word1[0] == word2[0] and word1[1] == word2[1] and word1[2] == word2[2] and word1[3] != word2[3] and word1[4] == word2[4]):
           Dic_Edge[word1, word2] = 1
    if (word1[0] == word2[0] and word1[1] == word2[1] and word1[2] == word2[2] and word1[3] == word2[3] and word1[4] != word2[4]):
           Dic_Edge[word1, word2] = 1
    
```

```python
def difference(a, b):
    count = sum(1 for a, b in zip(a, b) if a != b)
    return count
```

```python
for i in range(len(a)):
    for j in range(len(a)):
#         print (i, j)
#         sum = sum + 1
#         compare_two_words(a[i], a[j])
        if difference(a[i], a[j]) == 1:
            Dic_Edge[a[i], a[j]] = 1
```
